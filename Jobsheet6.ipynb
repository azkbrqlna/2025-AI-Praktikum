#### 1.	Melakukan Visualisasi Data menggunakan Pie Chart

```python
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Pistachio', 'Mango', 'Strawberry')
votes = (12, 11, 4, 8, 7)

colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D','#D53032')

explode = (0, 0, 0, 0.1, 0)

plt.pie(votes,
        labels=flavors,
       autopct='%1.2f%%',
       colors=colors,
       explode = explode,
       shadow = True)
plt.show()
```

#### 2.	Melakukan Visualisasi Data menggunakan Bar Chart

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

countries = ('Argentina', 'Bolivia', 'Brazil', 'Chile',  'Colombia', 'Ecuador', 'Falkland Islands', 'French Guiana', 'Guyana', 'Paraguay', 'Peru','Suriname', 'Uruguay', 'Venezuela')

populations = (45076704, 11626410, 212162757, 19109629,50819826, 17579085, 3481, 287750, 785409, 7107305, 32880332,585169, 3470475, 28258770)

df = pd.DataFrame({
    'Country' : countries,
    'Population' : populations,
})

df.sort_values(by='Population', inplace = True)
x_coords = np.arange(len(countries))

plt.bar(x_coords, df['Population'], tick_label=df['Country'])
plt.xticks(rotation=90) #rotates text for x-axis labels
plt.ylabel('Population')
plt.title('South American Populations')
plt.show()
```
